```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on="Country Code",how="left")
final_df.head(2)
```

```python
## TO check Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
## Pie chart top 3 country that uses Zomatp
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.2f%%");
```

```python
final_df.columns
```

```python
final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size()
```

```python
final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index()
```

```python
ratings = final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```

## Obervations:

1. When Ratings are between 4.5 to 4.9 is Excellent
2. When Ratings are between 4.0 to 4.4 is Very Good
3. When Ratings are between 3.5 to 3.9 is Good
4. When Ratings are between 3.0 to 3.4 is Average
5. When Ratings are between 2.5 to 2.9 is Average
6. When Ratings are between 2.0 to 2.4 is Poor

```python
ratings.head()
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings);
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings);
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue="Rating color");
```

```python
## Mapping Colors
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue="Rating color",palette=["blue","red","orange","yellow","green","green"]);
```

##Observations:

1. Not Rated count is very high
2. maximum number of rating are between 2.5 to 4.0

```python
#count pot it shows frequency of data
plt.figure(figsize=(12,6))
sns.countplot(x="Rating color",data=ratings,palette=["blue","red","orange","yellow","green","green"]);
```
